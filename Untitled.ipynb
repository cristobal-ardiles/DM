```python
# Obtendremos la superficie de √°reas verdes correspondientes a plazas o parques
df3 = df[["comuna", "area_m2", "estado_2","clase"]]
df3.head()
```

```python
grouped = df3.groupby("comuna").sum()
grouped = grouped.sort_values("area_m2", ascending=False)
grouped.to_csv("sup_total_areas_verdes.csv", index=False)
grouped
```
