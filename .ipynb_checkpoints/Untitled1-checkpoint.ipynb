```python
import pandas as pd
import numpy as np
```

# Lider

```python
df = pd.read_csv("webscraping/lider.csv")
df["Nombre"] = [x.strip(' ').strip('\n').strip(' ') for x in df["Nombre"]]
df["Direccion"] = [x.split(',')[0] for x in df["Direccion"]]
df_lider = df[["Nombre", "Direccion", "Comuna"]]
```

# Jumbo


```python
df_jumbo = pd.read_csv("webscraping/jumbo.csv")
df_jumbo[["Direccion","Comuna"]] = [x.split(',')[0:2] for x in df_jumbo["Direccion"]]
df_jumbo
```

# Unimarc

```python
df_unimarc = pd.read_csv("webscraping/unimarc.csv")
df_unimarc["Direccion"] = [x.strip("Dirección: ").split(',')[0] for x in df_unimarc["Direccion"]]
df_unimarc

comunas = df_unimarc["Comuna"].unique()
comunas = np.sort(comunas)
comunas

df_unimarc["Comuna"] = [x.capitalize() for x in df_unimarc["Comuna"]]
df_unimarc["Comuna"].unique()
```

```python
df_sta_isabel = pd.read_csv("webscraping/sta_isabel.csv")
df_sta_isabel[["Direccion", "Comuna"]] = [x.split(',')[0:2] for x in df_sta_isabel["Direccion"]]
df_sta_isabel

comunas_2 = np.sort(df_sta_isabel["Comuna"].unique())
comunas_2
```

```python
df_super = pd.concat([df_lider, df_jumbo, df_sta_isabel, df_unimarc], ignore_index=True)
df_super
```

# Tottus


```python
def parse(x):
    if '/' in x:
        return x.split('/')
    else:
        return x.split('-')
    
def parse_2(x):
    y = x.split(',')
    n = len(y)
    if n == 1:
        return [y[0], np.nan]
    elif n==2:
        return y
    else:
        return [y[0], build_string(y[1:])]
    
def build_string(list):
    ret = ""
    for elem in list:
        ret += (elem+',')
    return ret[:-1]
```

```python
df_tottus = pd.read_csv("webscraping/tottus.csv")
df_tottus = df_tottus[1:]
df_tottus[["Nombre","Direccion"]] = [parse(x) for x in df_tottus["Nombre"]]
df_tottus["Direccion"] = [x.replace("\n",'') for x in df_tottus["Direccion"]]
df_tottus[["Direccion", "Comuna"]] = [parse_2(x) for x in df_tottus["Direccion"]]
df_tottus
```

```python
# Como se puede ver, hay algunos valores incorrectos
df_tottus.at[23,"Direccion"] = "Caletera, Los Aromos 0441"
df_tottus.at[23,"Comuna"] = "Talagante"
df_tottus
```

```python
df_tottus.at[31, "Comuna"] = "Vitacura"
df_tottus.at[32, "Comuna"] = "Vitacura"
df_tottus
```

```python
df_tottus.at[13, "Comuna"] = "La Florida"
df_tottus.at[17, "Comuna"] = "Puente Alto"
df_tottus
```

```python
df_tottus["Comuna"] = [x.strip(" ") for x in df_tottus["Comuna"]]
df_tottus
```

```python
df_super_2 = pd.concat([df_super, df_tottus], ignore_index=True)
df_super_2.to_csv("supermercados.csv")
```
